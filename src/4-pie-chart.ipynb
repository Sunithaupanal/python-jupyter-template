### Exercise 4: Pie Chart with Percentages

```python
# TODO: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes  = [40, 25, 20, 15]

# Check if labels and sizes have the same length
if len(labels) != len(sizes):
    print("Error: labels and sizes must have the same length for the pie chart.")
else:
    # Create a Pie plot
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')

    # Title
    plt.title('Programming Language Pie Chart')

    # Show the plot
    plt.show()
```
