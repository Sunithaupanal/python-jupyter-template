### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Data
x = [6, 7, 8, 9, 10]
y = [36, 49, 64, 81, 100]
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 4, 9]
labels = ['C', 'C++', 'Python', 'Java']
sizes = [10, 9, 6, 5]

# Check if x and y have the same length
if len(x) != len(y):
    print("Error: x and y must have the same length for line and scatter plots.")
else:
    # Create a 2x2 grid of subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # First subplot: Line plot
    axs[0, 0].plot(x, y)
    axs[0, 0].set_title('Line Plot')

    # Second subplot: Bar plot (Check if categories and values have the same length)
    if len(categories) != len(values):
        print("Error: categories and values must have the same length for the bar plot.")
    else:
        axs[0, 1].bar(categories, values)
        axs[0, 1].set_title('Bar Plot')

    # Third subplot: Scatter plot
    axs[1, 0].scatter(x, y)
    axs[1, 0].set_title('Scatter Plot')

    # Fourth subplot: Pie chart (Check if labels and sizes have the same length)
    if len(labels) != len(sizes):
        print("Error: labels and sizes must have the same length for the pie chart.")
    else:
        axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
        axs[1, 1].set_title('Pie Chart')

    # Show the plots
    plt.show()
```
